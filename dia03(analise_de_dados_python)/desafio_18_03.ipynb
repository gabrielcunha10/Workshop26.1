```python
import numpy as np
import pandas as pd

dias_semana = [
    "Segunda",
    "Terça",
    "Quarta",
    "Quinta",
    "Sexta",
    "Sábado",
    "Domingo"
]

df = pd.DataFrame({
    "Dia da semana": dias_semana,
    "Temperatura": temperaturas
})

media_df = df['Temperatura'].mean()
max_df = df['Temperatura'].max()
min_df = df['Temperatura'].min()
print(f"Média das temperaturas da semana: {media_df:.2f}\n", f"Temperatura máxima da semana: {max_df:.2f}\n", f"Temperatura mínima da semana: {min_df:.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dias_semana = [
    "Segunda",
    "Terça",
    "Quarta",
    "Quinta",
    "Sexta",
    "Sábado",
    "Domingo"
]

df = pd.DataFrame({
    "Dia da semana": dias_semana,
    "Temperatura": temperaturas
})

x = df["Dia da semana"]
y = df["Temperatura"]

# Primeiro gráfico

plt.bar(x, y, label="Temperatura")
plt.xlabel("Dia da semana")
plt.ylabel("Temperatura")
plt.title("Temperatura de cada dia da semana - Gráfico de barras")
plt.grid()
plt.legend()
plt.show()

# Segundo gráfico

plt.plot(x, y, label="Temperatura")
plt.xlabel("Dia da semana")
plt.ylabel("Temperatura")
plt.title("Temperatura de cada dia da semana - Gráfico de linhas")
plt.grid()
plt.legend()
plt.show()

# Terceiro gráfico

plt.scatter(x, y, label="Temperatura")
plt.xlabel("Dia da semana")
plt.ylabel("Temperatura")
plt.title("Temperatura de cada dia da semana - Gráfico de dispersão")
plt.grid()
plt.legend()
plt.show()
```
